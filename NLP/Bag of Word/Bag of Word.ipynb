```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
```

```python
df = df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis=1)
df
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
df
```

```python
X = df.drop(['Category'], axis = 1)
Y = df.Category
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.Message, df.spam, test_size=0.2, random_state=42)
```

```python
X_train_list = X_train.tolist()
```

```python
v = CountVectorizer()

X_Train_cv = v.fit_transform(X_train)
X_Text_cv = v.transform(X_test)
```

```python
X_Train_cv
```

```python
X_Text_cv
```

```python
X_Train_cv.toarray()
```

```python
v.get_feature_names()[5]
```

```python
v.vocabulary_
```

```python
X_Train_cv_to_array = X_Train_cv.toarray()
```

```python
X_Train_cv_to_array[:5]
```

```python
model = MultinomialNB()
model.fit(X_Train_cv, Y_train)
```

```python
y_pred = model.predict(X_Text_cv)
```

```python
print(classification_report(y_pred, Y_test))
```
