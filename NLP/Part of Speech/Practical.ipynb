```python
import spacy as sp
```

```python
nlp = sp.load('en_core_web_sm')
```

```python
with open ('news_story.txt') as f:
    doc = f.read()
```

```python
doc = nlp(doc)
doc
```

```python
for tk in doc:
    print(tk, ' | ', tk.pos_)
```

```python
noun_list = []
number_list = []
for tk in doc:
    if tk.pos_ == 'NOUN':
        noun_list.append(tk)
    if tk.pos_ == 'NUM':
        number_list.append(tk)
```

```python
print('Nouns:-  ', noun_list)
print(" ")
print(" ")
print(" ")
print('Numbers:-  ', number_list)
```
