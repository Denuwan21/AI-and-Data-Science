```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state= 3)
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model.fit(X_train, Y_train, epochs=100)
```

```python
model.evaluate(X_test, Y_test)
```

```python
Y_pred = model.predict(X_test).reshape(-1)
print(Y_pred[:10])

Y_pred = np.round(Y_pred)
print(Y_pred[:10])

print(Y_test[:10])
```

```python
print(classification_report(Y_pred, Y_test))
```

```python
model_L = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model_L.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model_L.fit(X_train, Y_train, epochs=100)
```

```python
model_L.evaluate(X_test, Y_test)
```

```python
Y_predict = model_L.predict(X_test).reshape(-1)
Y_predict = np.round(Y_predict)

print(classification_report(Y_predict, Y_test))
```
